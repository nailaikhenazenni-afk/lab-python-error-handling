````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
````

````python
def calculate_total_price(customer_orders): 
   total_price = 0.0
   for product in customer_orders:
        valid_input = False
        while not valid_input:
            try:
                entry = input(f"Enter the price of {product}: ")
                price = float(entry)
                if price < 0:
                  raise ValueError("Price cannot be negative. Please enter a valid price.")
                print(f'Error: {price} cannot be negative. Please enter a valid price.')
                total_price += price
                valid_input = True
            except ValueError:
             print(f"Error: '{entry}' is not a valid number. Please try again.")
   return total_price
````

````python
def get_customer_orders(inventory: dict[str, int]) -> dict[str, int]:
    # 1. Solicitar la cantidad total de pedidos
    while True:
        try:
            orders_count = int(input("Enter the number of customer orders: "))
            if orders_count < 0:
                raise ValueError("The number of orders cannot be negative.")
            break
        except ValueError as e:
            print(f"Error: {e} Please enter a valid integer.")

    customer_orders = {}

    # 2. Recopilar cada producto
    for i in range(orders_count):
        while True:
            try:
                product_name = input(f"Gather the name for product {i + 1}: ")

                # Verificación 1: ¿Existe en el catálogo?
                if product_name not in inventory:
                    raise ValueError(f"'{product_name}' does not exist in our catalog.")
                
                # Verificación 2: ¿Hay stock disponible?
                # Consideramos cuántos ya ha pedido el usuario en este turno
                already_ordered = customer_orders.get(product_name, 0)
                if inventory[product_name] <= already_ordered:
                    raise ValueError(f"'{product_name}' is currently out of stock or insufficient units.")

                # Si pasa las validaciones, lo añadimos al "carrito" (diccionario)
                customer_orders[product_name] = already_ordered + 1
                print(f"Added {product_name} to your order.")
                break
                
            except ValueError as e:
                print(f"Invalid input: {e} Please try again.")

    return customer_orders

                

   
````
